```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import numpy as np
```

```python
x=[0.95,0.9,0.7,0.68,0.4,.3,0.2,0.05]

for i in range(8):
  plt.scatter((i+1),x[i], marker='^', color='Blue', s=80)
  plt.text((i+1), x[i], str(x[i]), ha='left', va='bottom',fontsize=12)


plt.ylim(0, 1)
plt.xlabel('Filter Index', fontsize=20)
plt.ylabel('Normalized Score', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```
